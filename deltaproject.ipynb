```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy as s
import seaborn as sns
df=pd.read_csv(r'C:\data.csv')
res_accgen_mh = s.stats.pearsonr(df['acceptance_genLSR'], df['MH_diff'])
print(res_accgen_mh) #super
res_accgen_cultrep = s.stats.pearsonr(df['acceptance_genLSR'], df['cult_rep'])
print(res_accgen_cultrep)
res_cultrep_cultfoodrep = s.stats.pearsonr(df['cult_rep'], df['cultfood_rep'])
print(res_cultrep_cultfoodrep)
res_extentdisc_catgdisc = s.stats.pearsonr(df['extent_disc'],df['catg_disc_scale'])
print(res_extentdisc_catgdisc) #super 
res_extentdisc_disceffgra = s.stats.pearsonr(df['disc_effect_grades'], df['extent_disc'])
print(res_extentdisc_disceffgra) #super
res_catgdisc_disceffgra = s.stats.pearsonr(df['disc_effect_grades'], df['catg_disc_scale'])
print(res_catgdisc_disceffgra)
res_catgdisc_catgjud = s.stats.pearsonr(df['catg_disc_scale'], df['catg_judgement'])
print(res_catgdisc_catgjud)

correlation = df.corr(method='pearson', numeric_only=True)
sns.heatmap(correlation)
```
